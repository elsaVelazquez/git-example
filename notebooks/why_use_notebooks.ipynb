```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# Fixing random state for reproducibility
np.random.seed(42)

size = 1000

x_purple = np.random.normal(loc=6, scale=1, size=size)
y_purple = np.random.normal(loc=6, scale=1, size=size)

x_orange = np.random.normal(loc=3, scale=1, size=size)
y_orange = np.random.uniform(high=3, size=size) - 2

x_green = np.random.normal(loc=1, scale=0.6, size=size)
y_green = np.random.normal(loc=4, scale=2, size=size)
```

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy, color='tab:blue', alpha=1):
    # no labels
    ax_histx.set_xticks([])
    ax_histx.set_yticks([])
    
    ax_histy.set_xticks([])
    ax_histy.set_yticks([])

    # the scatter plot:
    ax.scatter(x, y, color=color, alpha=alpha)
    
    # now determine nice bins by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth
    bins = np.arange(-lim, lim + binwidth, binwidth)
    
    # the histograms
    ax_histx.hist(x, bins=bins, color=color, alpha=0.6)
    ax_histy.hist(y, bins=bins, color=color, alpha=0.6, orientation='horizontal')
```

```python
# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# make lims
x_min, x_max = -2, 10
y_min, y_max = -3, 10

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(x_purple, y_purple, ax, ax_histx, ax_histy, color='tab:purple', alpha=0.2)
scatter_hist(x_orange, y_orange, ax, ax_histx, ax_histy, color='tab:orange', alpha=0.2)
scatter_hist(x_green, y_green, ax, ax_histx, ax_histy, color='tab:green', alpha=0.2)

# set xlim
ax.set_xlim(left=x_min, right=x_max)
ax_histx.set_xlim(left=x_min, right=x_max)

# set ylim
ax.set_ylim(bottom=y_min, top=y_max)
ax_histy.set_ylim(bottom=y_min, top=y_max)

plt.show()
```
